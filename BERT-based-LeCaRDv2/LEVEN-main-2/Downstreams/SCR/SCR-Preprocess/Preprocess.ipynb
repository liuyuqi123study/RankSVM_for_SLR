```python
for fn in os.listdir("/content/candidate/candidate_55192"):
  cand=json.load(open(os.path.join("/content/candidate/candidate_55192",fn),"r"))
  str1=cand['fact']
  txt_dict={}
  with open(os.path.join("/content/candidate/candidate_fact_0608",fn),"w",encoding="utf-8") as f:
    txt_dict["fact_part"]=str1
    txt_dict["pid"]=cand["pid"]
    f.write(json.dumps(txt_dict,ensure_ascii=False))

```

```python
data=list(jsonlines.open("train_query.json"))
with jsonlines.open("train_query_fact.json","w") as f:
  for i in data:
    txt_dict={}
    tmp=i["fact"]
    txt_dict["fact_part"]=tmp
    txt_dict["id"]=i["id"]
    jsonlines.Writer.write(f,txt_dict)
```
